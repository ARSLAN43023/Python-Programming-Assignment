# Question No: 1

Define Object Oriented Programming?

Answer:<br /> An object-oriented programming is a high level programming,language based on the object-oriented model. It is a paradigm based on 
the concept of objects, which contain data in the form of fields 
(known as attributes)and in the form of procedures (known as methods).In OOP, computer programs are designed by making them out of objects that interact with one another. Many modern programming languages are object-oriented,however some older programming languages, such as Pascal, do offer object-oriented versions.<br /> Examples of object-oriented programming 
languages include Java, C++ and Python.

Differentiate between function and method?<br />
Function:<br />
A function is a block of code to carry out a specific task, will contain 
its own scope and is called by name. All functions may contain zero (no) 
arguments or more than one arguments. On exit, a function can or cannot 
return one or more values.<br />

Method:<br />
A method in python is somewhat similar to a function, except it is associated with object/classes. Methods in python are very similar to functions except for two major differences.
•	The method is implicitly used for an object for which it is called.
•	The method is accessible to data that is contained within the class.

```python
class Car:
    # 5 attributes of class car      
    def __init__(self,model,color,name,cartype,horsepower):    
        self.model=model;
        self.color=color;
        self.name=name;
        self.cartype= cartype;
        self.horsepower=horsepower;

    # Method 1
    def Method_1(self):    
        print("**************\n1st Car Details\n*****************\nModel:",self.model,"\nColor:",self.color,"\nName:",self.name,"\nType:",self.cartype,"\nHorsePower:",self.horsepower)
    # Method 2
    def Method_2(self):    
        print("**************\n2nd Car Details\n********************\nModel:",self.model,"\nColor:",self.color,"\nName:",self.name,"\nType:",self.cartype,"\nHorsePower:",self.horsepower)
    # Method 3
    def Method_3(self):    
        print("*******************\n3rd Car Details\n**************** \nModel:",self.model,"\nColor:",self.color,"\nName:",self.name,"\nType:",self.cartype,"\nHorsePower:",self.horsepower)    

# First Object Instance of Class
CarObj1 = Car("2019","Grey","FAW V2","Hatch Back","1300 CC")    
CarObj1.Method_1()

# Second Object Instance of Class
CarObj2 = Car("2016","Golden","Suzuki APV","SUV","1500 CC")    
CarObj2.Method_2()  

# Third Object Instance of Class
CarObj3 = Car("2010","Green","Toyota Corolla","5 Seater","1300 CC")    
CarObj3.Method_3()  

# Fourth Object Instance of Class
CarObj4 = Car("1998","Grey","Daewoo Racer","5 Seater","1500 CC")      

# Fifth Object Instance of Class
CarObj5 = Car("2017","Black","Toyota Land Cruiser","Jeep","2500 CC")  


print("****************\n4th Car Details\n**************\nModel:",CarObj4.model,"\nColor:",CarObj4.color,"\nName:",CarObj4.name,"\nType:",CarObj4.cartype,"\nHorsePower:",CarObj4.horsepower)    

print("****************\n5th Car Details\n**************\nModel:",CarObj5.model,"\nColor:",CarObj5.color,"\nName:",CarObj5.name,"\nType:",CarObj5.cartype,"\nHorsePower:",CarObj5.horsepower)    
```
